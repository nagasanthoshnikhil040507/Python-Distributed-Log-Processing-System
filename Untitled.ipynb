```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("TestApp") \
    .getOrCreate()

sc = spark.sparkContext  # now sc is defined

data = [1, 2, 3, 4, 5]
rdd = sc.parallelize(data)

print(rdd.collect())
rdd2 = rdd.map(lambda x: x * 2)
result = rdd2.collect()
print(result)
```
