```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("FastLoad") \
    .master("local[*]") \
    .getOrCreate()

df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv("../Data/raw_logs/logdata.csv")

df.show(5, truncate=False)
```
