```python
from pyspark.sql import SparkSession

# If Spark session is not already running, start it
spark = SparkSession.builder \
    .appName("BasicTest2") \
    .master("local[1]") \
    .getOrCreate()

# Sample data
data = [
    (1, "Alice", 25),
    (2, "Bob", 30),
    (3, "Charlie", 22),
    (4, "David", 28)
]

# Column names
columns = ["id", "name", "age"]

# Create DataFrame
df = spark.createDataFrame(data, columns)

# Show the DataFrame
print("Original DataFrame:")
df.show()

# Filter: only people older than 25
print("People older than 25:")
df.filter(df.age > 25).show()

# Add a new column: age in 5 years
from pyspark.sql.functions import col
df2 = df.withColumn("age_plus_5", col("age") + 5)
print("DataFrame with age_plus_5 column:")
df2.show()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Start Spark session
spark = SparkSession.builder \
    .appName("BasicTest") \
    .master("local[1]") \
    .getOrCreate()

# Sample data
data = [
    (1, "Alice", 25),
    (2, "Bob", 30),
    (3, "Charlie", 22),
    (4, "David", 28),
    (5, "Eva", 35)
]

columns = ["id", "name", "age"]

# Create DataFrame
df = spark.createDataFrame(data, columns)

# Show full DataFrame
print("Full DataFrame:")
df.show()

# Count rows
print("Number of rows:", df.count())

# Show first row
print("First row:")
print(df.first())

# Take first 3 rows
print("First 3 rows using take():")
print(df.take(3))

# Filter: people older than 28
print("People older than 28:")
df.filter(col("age") > 28).show()

# Loop through DataFrame safely
print("Looping through rows:")
for row in df.collect():
    print(row)

# Stop Spark
spark.stop()
print("Spark session stopped")
```
